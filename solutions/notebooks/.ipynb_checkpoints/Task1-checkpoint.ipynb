# **PREFACE** <a class="anchor" id="first-bullet"></a>

<span class="c5">Although the plot specifies the x-axis to be in nanometers, I have taken the unit to be in Angstrom (</span><span class="c19 c33">Å)</span><span class="c5 c8"> instead. My reasoning for this is as follows:</span>

**<span class="c5">Assumption :</span><span class="c8 c37 c43"> We're dealing with the gas in vacuum, and not air.</span>**

<span class="c8 c5">Using the Rydberg formula for calculating the wavelengths corresponding to the transitions of Hydrogen, we note that:</span>

<span class="c5">(a) The shortest wavelength possible is approximately 911.7</span> <span class="c0">Å.</span>

<span class="c5">(b) The longest wavelength possible is approximately 1215.7</span> <span class="c0">Å.</span>

<span class="c5">Ignoring the unit for now, we can see that the value from (a) corresponds to the "Plunge" at around 900, and the dip at around 1220 also lines up with the value from (b). This leads me to believe that the true unit for the x-axis might not be in nm. The reference cited in the question (</span><span class="c18 c19">[https://physics.nist.gov/PhysRefData/Handbook/Tables/hydrogentable2.htm](https://www.google.com/url?q=https://physics.nist.gov/PhysRefData/Handbook/Tables/hydrogentable2.htm&sa=D&source=editors&ust=1616619658701000&usg=AOvVaw2PETS8nukkv-GbAi98HvI1)</span><span class="c5">) also appears to corroborate this, as the units provided in the tables there are also in</span> <span class="c19">Å</span><span class="c8 c5"> and not nm.</span>

<span class="c5">For Hydrogen to have transitions in the nm range, the transitions would have to correspond to the Paschen series. But the lower limit for wavelengths of Hydrogen in the series is at around 821 nm, which despite being present on the x axis, does not show any corresponding changes in the y-axis. This observation also is in agreement with the assumption that the unit is probably not nm, but rather</span> <span class="c19">Å</span><span class="c8 c5"> instead.</span>

<span class="c19">Hence, I have assumed the true unit of the x-axis to be in Å and not nm for this task (both parts A and B). This assumption has also been verified and confirmed by Shravan Choudhari.</span>

<span class="c37">Note :</span> <span class="c11">We can assume the absorption and emission spectra to be “inverses” of each other, because the wavelength at which both phenomena occur is the same, and the processes which produce the radiations are just inverses of each other. In the case of absorption, the electron absorbs the incident photon and “jumps” to a higher energy level (assuming the intensity of the incident photo is greater than the required minimum threshold intensity), while emission deals with an electron “falling” from a higher energy state to a lower one, while releasing a photon in the process.</span>

<span class="c27">Also, for each required minimum incident wavelength intensity for emission, we note that the observed intensity of light at that wavelength has to be less than or equal to 1 - I</span><span class="c16">req</span><span class="c27"> (I</span><span class="c16">req</span><span class="c11"> being the required incident intensity of radiation) for the element to definitely absorb the radiation.</span>

1.  <span class="c30">How are the lines in ‘C’ related?</span> <a class="anchor" id="c-bullet"></a>

<span class="c13">From the Equation (1) (Rydberg’s formula) and the explanation of the first line of ‘C’ from above, we know that the line at 911.7 is predicted from the lower limit of the wavelength of radiation for the Lyman series. The other lines in ‘C’ can be accounted for by substituting the other various values of</span> ![](images/image9.png)<span class="c13">in the formula. Upon comparison with the</span> <span class="c3">[emission spectrum of Hydrogen](https://www.google.com/url?q=https://upload.wikimedia.org/wikipedia/commons/7/76/LymanSeries.svg&sa=D&source=editors&ust=1616619658715000&usg=AOvVaw3eXb7xWb5_mBIzaOJBMCeb)</span><span class="c13">, we can see that the emission lines in the image line up with the absorption spectrum. Therefore, we can assume (by correlation) that the lines in ‘C’ must belong to the element Hydrogen. Hence, we can say that the lines in ‘C’ are related,</span> <span class="c25 c13">because all of them belong to the same element.</span>

2.  <span class="c30">What causes the “Plunge” near 900?</span> <a class="anchor" id="900-bullet"></a>

<span class="c13">The plunge near 900 (at 911.7 Å to be precise) is caused due to the radiations approaching the limit of the Lyman series for Hydrogen.</span> <span class="c13 c25">As n</span><span class="c25 c13 c41">2</span><span class="c25 c13"> approaches infinity, we end up with multiple lines appearing to converge at this wavelength</span><span class="c2">, which causes all radiations in these frequencies to be absorbed by the gas. Hence, we observe a “plunge” here as there is no light making it past the gas cloud to the observer at these frequencies.</span>
